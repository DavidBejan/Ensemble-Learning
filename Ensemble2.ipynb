```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def function(w):
    total = 0
    weak_class = 5 + int(w/2) + 1
    for n in range(weak_class, 11):
        total += (np.math.factorial(10) / (np.math.factorial(n)*np.math.factorial(10-n))) * 0.6**n * 0.4**(10-n) * 0.15
    weak_class = 4-int(w/2)+1
    for n in range (weak_class, 11):
        total += (np.math.factorial(10) / (np.math.factorial(n)*np.math.factorial(10-n))) * 0.6**n * 0.4**(10-n) * 0.75
    return total
```
